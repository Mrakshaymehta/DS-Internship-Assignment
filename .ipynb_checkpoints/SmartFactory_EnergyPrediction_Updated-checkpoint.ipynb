# Smart Factory Energy Prediction Challenge
This notebook includes EDA, preprocessing, feature selection, and regression model building to predict `equipment_energy_consumption` using sensor data from a manufacturing facility.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler

# Load dataset
df = pd.read_csv('data.csv')

# Clean column names
df.columns = df.columns.str.strip()

# Convert target column to numeric
df['equipment_energy_consumption'] = pd.to_numeric(df['equipment_energy_consumption'], errors='coerce')
```

## Exploratory Data Analysis

```python
# Check for missing values
print("Missing values:")
print(df.isnull().sum())

# Describe basic statistics
df.describe()
```

## Correlation with Target Variable

```python
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(12,6))
sns.heatmap(correlation, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Show top correlated features with the target
correlation["equipment_energy_consumption"].sort_values(ascending=False).head(10)
```

## Data Preprocessing

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Replace all non-numeric error placeholders with NaN
df.replace(['???', 'error', 'na', 'NA', '--', 'null'], np.nan, inplace=True)

# Drop rows where target is missing
df.dropna(subset=['equipment_energy_consumption'], inplace=True)

# Fill other missing values with median
df.fillna(df.median(numeric_only=True), inplace=True)

# Drop non-numeric and timestamp columns from features
X = df.drop(columns=['timestamp', 'equipment_energy_consumption'])
y = df['equipment_energy_consumption']

# Make sure all columns are numeric
X = X.apply(pd.to_numeric, errors='coerce')

# Scale
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
df.to_csv('Processed Data.csv')
```

## Model Training and Evaluation

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```

## Key Insights and Recommendations

- The model identifies patterns between zone temperatures and energy consumption, though correlations are weak.
- Data quality and additional contextual variables may improve predictive power.
- Zones with consistent high consumption may benefit from temperature regulation, humidity control, or scheduling optimization.

## ⚠ Outlier Detection using IQR Method

```python
import pandas as pd

# Ensure the feature is numeric
df['zone1_temperature'] = pd.to_numeric(df['zone1_temperature'], errors='coerce')

# Drop NaNs introduced during conversion (optional, depending on your use case)
df = df.dropna(subset=['zone1_temperature'])

# Outlier detection using IQR
feature = 'zone1_temperature'
Q1 = df[feature].quantile(0.25)
Q3 = df[feature].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers
outliers = df[(df[feature] < Q1 - 1.5 * IQR) | (df[feature] > Q3 + 1.5 * IQR)]
print(f"Number of outliers in {feature}: {len(outliers)}")
```

## 📊 Feature Importance Visualization

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

# Drop non-numeric columns like timestamp (or others, as needed)
X = df.drop(columns=['equipment_energy_consumption'])
X = X.select_dtypes(include=[np.number])  # keep only numeric columns
y = df['equipment_energy_consumption']

# Train model
model = RandomForestRegressor()
model.fit(X, y)

# Feature importance
importances = model.feature_importances_
indices = np.argsort(importances)[::-1]
features = X.columns[indices]

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x=features, y=importances[indices])
plt.xticks(rotation=90)
plt.title("Feature Importances")
plt.tight_layout()
plt.show()
```

## 📈 Scatter Plot and Residual Analysis

```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_pred = model.predict(X)
residuals = y - y_pred

# Scatter plot
plt.figure(figsize=(10, 5))
sns.scatterplot(x=y, y=y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted")
plt.grid(True)
plt.show()

# Residual plot
plt.figure(figsize=(10, 5))
sns.histplot(residuals, bins=30, kde=True)
plt.title("Residual Distribution")
plt.xlabel("Residuals")
plt.show()
```
